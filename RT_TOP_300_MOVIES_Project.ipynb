```python
from re import M
Movie_rank = []
Movie_title = []
Movie_rate = []
Movie_cast = []
Movie_director = []
Movie_release_year = []
Movie_synopsis = []
Movie_critics_consensus = []
Movie_link = []
Movie_poster = []
for page in range(1,3) :
    response = requests.get(f"https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/{page}/")
    soup = BeautifulSoup(response.text, "html.parser")
    Top_movies = soup.find_all("div", class_= "row countdown-item")
    for movie in Top_movies :
        Movie_rank.append(movie.find("div", class_="countdown-index-resposive").get_text().replace("#" , ""))
        Movie_title.append(movie.find("h2").find("a").get_text())
        Movie_link.append(movie.find("a").attrs["href"])
        Movie_rate.append(movie.find("span", class_="tMeterScore").get_text())
        Movie_cast.append(movie.find("div", class_="info cast").get_text().strip().replace("Starring: " , ""))
        Movie_director.append(movie.find("div", class_="info director").get_text().strip().replace("Directed By: " , ""))
        Movie_release_year.append(movie.find("span", class_="subtle start-year").get_text().replace("(" , "").replace(")" , ""))
        Movie_synopsis.append(movie.find("div", class_="info synopsis").get_text().strip().replace("Synopsis: " , "").replace("\n               [More]" , "" ))
        Movie_critics_consensus.append(movie.find("div", class_="info critics-consensus").get_text().strip().replace("Critics Consensus: " , ""))
        Movie_poster.append(movie.find("img" , class_ = "article_poster").attrs["src"])
```

```python
with open("Top_movies.csv" , "w" , newline = "") as file :
    writer = csv.DictWriter(file , fieldnames = ["Rank" , "Title" , "Rate" , "Cast" , "Director" , "Release_year" , "Synopsis" , "Critics_consensus" , "Poster"])
    writer.writeheader()
    for i in range(len(Movie_rank)) :
        writer.writerow({"Rank" : Movie_rank[i] , "Title" : Movie_title[i] , "Rate" : Movie_rate[i] , "Cast" : Movie_cast[i] , "Director" : Movie_director[i] , "Release_year" : Movie_release_year[i] , "Synopsis" : Movie_synopsis[i] , "Critics_consensus" : Movie_critics_consensus[i] , "Poster" : Movie_poster[i]})
```
